# PIE CHART

BASED ON RATING GIVING A REVIEW TO THE MOVIES

```python
def Movie_review(rating):
    if float(rating) >= 8:
        return "Very Good"
    elif float(rating) >= 7 and rating < 8:
        return "Good"
    elif float(rating) >= 6 and rating < 7:
        return "Neutral"
    else:
        return "Bad"
```

```python
df["Review"]= df["rating"].apply(lambda x : Movie_review(x))
```

# VECTORISATION

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
features = ['description', 'genre', 'stars', 'directors']
```

```python
def combine_features(row):
    return row['description']+' '+row['genre']+' '+row['directors']+' '+row['stars']
```

```python

df['combined_features'] = df.apply(combine_features, axis = 1)
df.reset_index(drop=True,inplace=True)
df
```

```python
new_df = df[['movie_name','combined_features']]
new_df.head()
```

```python
print(new_df.loc[0, 'combined_features'])
```

```python
cv = CountVectorizer()
count_matrix = cv.fit_transform(new_df['combined_features'])
```

```python
count_matrix
```

```python
cosine_sim = cosine_similarity(count_matrix)
cosine_sim
```

```python
new_df['index'] = new_df.index
new_df
```

```python
def get_movie_name_from_index(index):
    return new_df[new_df.index == index]["movie_name"].values[0]
def get_index_from_movie_name(movie_name):
    return new_df[new_df.movie_name == movie_name]["index"].values[0]
```

```python
movie_user_likes = "Gloria"
movie_index = get_index_from_movie_name(movie_user_likes)
similar_movies = list(enumerate(cosine_sim[movie_index]))
```

```python
sorted_similar_movies = sorted(similar_movies,key=lambda x:x[1],reverse=True)[1:]
```

```python
i=0
print("Top 10 similar movies to "+movie_user_likes+" are:\n")
for element in sorted_similar_movies:
    print(get_movie_name_from_index(element[0]))
    i=i+1
    if i>10:
        break
```

# TfidfVectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer


tf = TfidfVectorizer(analyzer = 'word', ngram_range = (1,1),stop_words = 'english')
tfidf_matrix = tf.fit_transform(new_df['combined_features']).toarray()
cosine_similar = cosine_similarity(tfidf_matrix, tfidf_matrix)
cosine_similar
```

```python
def get_movie_name_from_index(index):
    return new_df[new_df.index == index]["movie_name"].values[0]
def get_index_from_movie_name(movie_name):
    return new_df[new_df.movie_name == movie_name]["index"].values[0]
```

```python
movie_user_likes = "The Shawshank Redemption"
movie_index = get_index_from_movie_name(movie_user_likes)
similar_movies = list(enumerate(cosine_similar[movie_index]))
```

```python
sorted_similar_movies = sorted(similar_movies,key=lambda x:x[1],reverse=True)[1:]
```

```python
i=0
print("Top 10 similar movies to "+movie_user_likes+" are:\n") #\n is for printing names in vertically on different lines
for element in sorted_similar_movies:
    print(get_movie_name_from_index(element[0]))
    i=i+1
    if i>10:
        break
```

# Recomendation by KNN

```python
tfidf_matrix
```

```python
tfidf_matrix_df = pd.DataFrame(tfidf_matrix)
```

```python
tfidf_matrix_df.shape
```

```python
tfidf_matrix_df
```

```python
tfidf_matrix_df.index = df.movie_name
```

```python
tfidf_matrix_df
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
# Build NearestNeighbors Object
model_nn = NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=10)
```

```python
model_nn.fit(tfidf_matrix_df)
```

```python
# Get top 10 nearest neighbors 
indices=model_nn.kneighbors(tfidf_matrix_df.loc[['Gloria']], 10, return_distance=False)

# Print the recommended books
print("Recommended Books:")
for index, value in enumerate(tfidf_matrix_df.index[indices][0]):
    print((index+1),". ",value)
```
